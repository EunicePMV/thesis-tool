```python
import json

# Assuming you have a JSON file named 'data.json'
file_path = 'PSOreports.json'

# Load JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)

# Now 'data' contains the contents of the JSON file
print(json.dumps(data, indent=2))
```

```python
from IPython.display import JSON

# Display the prettified JSON data
JSON(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

json_data = data

# Extracting data from the JSON
folds = ["fold-1", "fold-2", "fold-3", "fold-4"]
classes = ["blb", "hlt", "rb", "sb"]

precision_data = {fold: {cls: data[cls]["precision"] for cls in classes} for fold, data in json_data.items()}
recall_data = {fold: {cls: data[cls]["recall"] for cls in classes} for fold, data in json_data.items()}
f1_data = {fold: {cls: data[cls]["f1-score"] for cls in classes} for fold, data in json_data.items()}

# Create a grouped bar plot for each class
for cls in classes:
    precision_values = [precision_data[fold][cls] for fold in folds]
    recall_values = [recall_data[fold][cls] for fold in folds]
    f1_values = [f1_data[fold][cls] for fold in folds]

    bar_width = 0.25
    index = np.arange(len(folds))

    plt.bar(index, precision_values, width=bar_width, label='Precision')
    plt.bar(index + bar_width, recall_values, width=bar_width, label='Recall')
    plt.bar(index + 2 * bar_width, f1_values, width=bar_width, label='F1-score')

    plt.xlabel('Folds')
    plt.ylabel('Metrics')
    plt.title(f'{cls} Metrics Across Folds')
    plt.xticks(index + bar_width, folds)
    plt.legend()
    plt.show()
```

```python
pso_metrics = data

accuracy_data = {fold: pso_metrics[fold]["accuracy"] for fold in folds}

# Plotting accuracy across folds
plt.figure(figsize=(10, 6))
plt.plot(folds, [accuracy_data[fold] for fold in folds], marker='o')
plt.title("Accuracy Across Folds")
plt.xlabel("Folds")
plt.ylabel("Accuracy")
plt.show()
plt.show()
```
